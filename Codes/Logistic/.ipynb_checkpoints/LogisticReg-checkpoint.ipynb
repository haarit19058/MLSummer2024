```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import pandas as pd

data = pd.read_csv("framingham.csv").iloc[100:1000]

columns_to_check = ['age','cigsPerDay','totChol']

data_cleaned = data.dropna(subset=columns_to_check)

data = data_cleaned[(data_cleaned[columns_to_check] == 0).any(axis=1)]

data
```

```python
rawAge = np.array(data["age"])
ageMean = rawAge.mean()
ageStd = rawAge.std()
standardAge = np.array((rawAge-ageMean)/ageStd)
standardAge[np.isnan(standardAge)]=0


male = np.array(data["male"])



rawCigsPerDay = np.array(data["cigsPerDay"])
cigsPerDayMean = rawCigsPerDay.mean()
cigsPerDayStd = rawCigsPerDay.std()
standardCigsPerDay = np.array((rawCigsPerDay-cigsPerDayMean)/cigsPerDayStd+1)
standardCigsPerDay[np.isnan(standardCigsPerDay)]=0


rawTotChol = np.array(data["totChol"])
totCholMean = rawTotChol.mean()
totCholStd = rawTotChol.std()
standardTotChol = np.array((rawTotChol-totCholMean)/totCholStd)
standardTotChol[np.isnan(standardTotChol)]=0

tenYearCHD = np.array(data["TenYearCHD"])
```

```python
def sigmoid(z):
    return (1/(1+np.exp(-z)))

def cost(x,y,w,b):
    m = x.shape[0]
    return (-1/m)*np.sum(y*np.log(sigmoid(np.dot(w,x)+b))+(1-y)*np.log(1-sigmoid(np.dot(w,x)+b)))

def gradient(x,y,w,b,alpha):
    m = x[0].shape[0]

    dj_dws = np.zeros(w.shape[0])
    for i in range(w.shape[0]):
        # print(sigmoid(np.dot(w,x)+b))
        # print(y)
        # print(x[i])
        dj_dws[i] = (1/m)*np.sum((sigmoid(np.dot(w,x)+b)-y)*x[i])

    dj_db = (1/m)*np.sum(sigmoid(np.dot(w,x))-y)

    w = w - alpha*dj_dws

    b = b - alpha*dj_db

    return w,b


```

```python
x = np.array([standardAge,male,standardCigsPerDay,standardTotChol])
y = tenYearCHD
w = np.zeros(x.shape[0])
b = 0

iterations = 100000

ws1 = np.zeros(iterations)
ws2 = np.zeros(iterations)
ws3 = np.zeros(iterations)
ws4 = np.zeros(iterations)
bs = np.zeros(iterations)
costs = np.zeros(iterations)

for i in range(iterations):
    try:
        w,b = gradient(x,y,w,b,0.0001)
        Cost = cost(x,y,w,b)
        ws1[i] = w[0]
        ws2[i] = w[1]
        ws3[i] = w[2]
        ws4[i] = w[3]
        bs[i] = b
        costs[i]=Cost
    except Exception as e:
        print(e)
print(w,b)

plt.plot(ws1,costs)
plt.show()
plt.plot(ws2,costs)
plt.show()
plt.plot(ws3,costs)
plt.show()
plt.plot(ws4,costs)
plt.show()
plt.plot(bs,costs)
plt.show()

```
