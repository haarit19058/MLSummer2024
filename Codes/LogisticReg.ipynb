```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plotfunc(X,Y):
    xyes1 = []
    xno1 = []
    xyes2 = []
    xno2 = []
    for i in range(len(X)):
        if Y[i]==1:
            xyes1.append(X[i][0])
            xyes2.append(X[i][1])
        else:
            xno1.append(X[i][0])
            xno2.append(X[i][1])
    plt.scatter(xyes1,xyes2,edgecolor="k",zorder=10)
    plt.scatter(xno1,xno2,edgecolor="k",zorder=10)

```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cost(x, y, w, b):
    m = x.shape[0]
    z = np.dot(x, w) + b
    return (-1 / m) * np.sum(y * np.log(sigmoid(z)) + (1 - y) * np.log(1 - sigmoid(z)))

def multiGradient(X, Y, w, b, alpha):
    m = X.shape[0]
    z = np.dot(X, w) + b
    h = sigmoid(z)
    error = h - Y
    
    dj_dw = (1 / m) * np.dot(X.T, error)
    dj_db = (1 / m) * np.sum(error)

    w -= alpha * dj_dw
    b -= alpha * dj_db

    return w, b

def computewb(x, y, iterations=100000, alpha=0.001):
    w = np.zeros(x.shape[1])
    b = 0
    
    for i in range(iterations):
        w, b = multiGradient(x, y, w, b, alpha)
    
    return w, b
```

```python
from sklearn.datasets import make_circles, make_classification, make_moons
from matplotlib.colors import ListedColormap
cm_bright = ListedColormap(["#FF0000", "#0000FF"])
circles_X , circles_Y= make_circles(noise=0.2, factor=0.5,random_state=1)
classify_X, classify_Y = make_classification(n_features=2, n_redundant=0, n_informative=2, random_state=0, n_clusters_per_class=1)
moons_X,moons_Y = make_moons(noise=0.3, random_state=1)

data = [
    (circles_X, circles_Y),
    (classify_X, classify_Y),
    (moons_X, moons_Y)
]

plt.figure(figsize=(14, 6))

for i, (X, Y) in enumerate(data, start=1):
    plt.subplot(1, 3, i)
    plotfunc(X, Y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),np.arange(y_min, y_max, 0.02))

    X_new = []
    for i in range(len(X)):
        X_new.append([X[i][0],X[i][0]**2,X[i][0]**3,X[i][1],X[i][1]**2,X[i][1]**3])
    X_new = np.array(X_new)
    # print(X_new[0],X[0])
    
    w,b = computewb(X_new,Y,alpha=0.1)
    print(w,b)
    xrav = xx.ravel()
    yrav = yy.ravel()
    
    z = np.dot(np.c_[xrav, xrav**2, xrav**3, yrav, yrav**2,yrav**3], w) + b
    # print(z)
    Z = sigmoid(z)
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.2, cmap=cm_bright)

plt.show()
```
