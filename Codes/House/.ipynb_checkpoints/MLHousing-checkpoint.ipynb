```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
data = pd.read_csv("Housing.csv")
data
```

```python
prices  = np.array(data['price']/1000000)
area = np.array(data['area']/1000)
bedrooms = np.array(data['bedrooms'])
```

```python
w = 0
b = 0

def gradientDescent(w,b,alp,x,y):
    m = len(x)
    
    dj_dw = (1/m)*np.sum(x*(w*x+b-y))
    dj_db = (1/m)*np.sum((w*x+b-y))

    w = w - alp * dj_dw
    b = b - alp * dj_db

    cf = (1/(2*m))*np.sum((w * x + b - y)**2)
    
    return cf,w,b

cfValues = []
w_values = []
b_values = []
for i in range(10000):
    cf,w,b = gradientDescent(w,b,0.02,area,prices)
    cfValues.append(cf)
    w_values.append(w)
    b_values.append(b)

print(w,b)
plt.plot(np.arange(10000),cfValues)
plt.show()
plt.plot(w_values,cfValues)
plt.show()
plt.plot(b_values,cfValues)
plt.show()
```

```python
plt.scatter(area,prices)
# b= 8
# w = 0.03
pr = w * area + b
plt.scatter(area,pr)


plt.show()
```

```python
# Assuming you have defined prices and area
m = prices.shape[0]
x = area
y = prices

# Define the range for w and b
w = np.arange(-100, 100)
b = np.arange(-100, 100)
w_, b_ = np.meshgrid(w, b)

# Initialize the cost array
cost = np.zeros(w_.shape)

# Calculate cost for each combination of w and b
for i in range(w_.shape[0]):
    for j in range(w_.shape[1]):
        cost[i][j] = (1/(2*m)) * np.sum((w_[i][j]*x + b_[i][j] - y)**2) 

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w_, b_, cost, cmap="viridis")
ax.set_title("Cost Function")
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("Cost")

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Housing.csv")
data
```

```python
m=100
n=110
prices  = np.array((data['price'][m:n]-data['price'][m:n].mean())/data['price'][m:n].std())
area = np.array((data['area'][m:n]-data["area"][m:n].mean())/data['area'][m:n].std())
```

```python
w = np.arange(-100,100)
b = np.arange(-100,100)

x = area
y = prices
m = x.shape[0]

w_ , b_ = np.meshgrid(w,b)
cf = np.zeros(w_.shape)

for i in range(w.shape[0]):
    for j in range(w.shape[0]):
        cf[i][j] = (1/(2*m)) * np.sum((w_[i][j]*x+b_[i][j]-y)**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w_, b_, cf, cmap="viridis")
ax.set_title("Cost Function")
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("Cost")

# Show the plot
plt.show()
```

```python
w = 0
b = 0

def gradientDescent(w,b,alp,x,y):
    m = len(x)
    
    dj_dw = (1/m)*np.sum(x*(w*x+b-y))
    dj_db = (1/m)*np.sum((w*x+b-y))

    w = w - alp * dj_dw
    b = b - alp * dj_db

    cf = (1/(2*m))*np.sum((w * x + b - y)**2)
    
    return cf,w,b

cfValues = []
w_values = []
b_values = []
for i in range(10000):
    cf,w,b = gradientDescent(w,b,0.02,area,prices)
    cfValues.append(cf)
    w_values.append(w)
    b_values.append(b)

print(w,b)
plt.plot(np.arange(10000),cfValues)
plt.show()
plt.plot(w_values,cfValues)
plt.show()
plt.plot(b_values,cfValues)
plt.show()
```
